```python
import pandas as pd
url = 'gesture0train.csv'
df0 = pd.read_csv(url)

url = 'gesture1train.csv'
df1 = pd.read_csv(url)

url = 'gesture2train.csv'
df2 = pd.read_csv(url)

url = 'gesture3train.csv'
df3 = pd.read_csv(url)

url = 'gesture4train.csv'
df4 = pd.read_csv(url)

url = 'gesture5train.csv'
df5 = pd.read_csv(url)

url = 'gesture6train.csv'
df6 = pd.read_csv(url)

url = 'gesture7train.csv'
df7 = pd.read_csv(url)

temp0 = df0.append(df1)
temp1 = temp0.append(df2)
temp2 = temp1.append(df3)
temp3 = temp2.append(df4)
temp4 = temp3.append(df5)
temp5 = temp4.append(df6)
temp6 = temp5.append(df7)

Y=temp3.iloc[:,-1:]
X=temp3.iloc[:,:-1]
```

```python
url = 'gesture0val.csv'
v0 = pd.read_csv(url)

url = 'gesture1val.csv'
v1 = pd.read_csv(url)

url = 'gesture2val.csv'
v2 = pd.read_csv(url)

url = 'gesture3val .csv'
v3 = pd.read_csv(url)

url = 'gesture4val.csv'
v4 = pd.read_csv(url)

url = 'gesture5val.csv'
v5 = pd.read_csv(url)

url = 'gesture6val.csv'
v6 = pd.read_csv(url)

url = 'gesture7val.csv'
v7 = pd.read_csv(url)

tempv0 = v0.append(v1)
tempv1 = tempv0.append(v2)
tempv2 = tempv1.append(v3)
tempv3 = tempv2.append(v4)
tempv4 = tempv3.append(v5)
tempv5 = tempv4.append(v6)
tempv6 = tempv5.append(v7)

Yval=tempv3.iloc[:,-1:]
Xval=tempv3.iloc[:,:-1]
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
estimator = ExtraTreesClassifier(n_estimators=100, max_depth=None, random_state=2)
estimator.fit(X, Y['target'])
predict = estimator.predict(Xval)
accuracy_score(Yval['target'],predict)
```
